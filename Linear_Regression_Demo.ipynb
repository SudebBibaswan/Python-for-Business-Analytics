```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("Housing_data.csv")
```

```python
data.head()
```

```python
data['parking_dum']=np.where(data['parking']=="Yes",1,0)
data['furnished_dum']=np.where(data['furnished']=="Yes",1,0)
```

```python
data.drop(['parking','furnished'],axis=1,inplace=True)
```

```python
cols_indep=set(data.columns)-set(['id','sale_price'])
```

```python
cols_indep
```

```python
X=data[list(cols_indep)]
Y=data['sale_price']
```

```python
X
```

```python
Y
```

```python
import statsmodels.api as sm
```

```python
X = sm.add_constant(X)
```

```python
X
```

```python
model = sm.OLS(Y,X)
```

```python
results = model.fit()
```

```python
results.summary()
```

```python
X=X.drop(['house_age','garden_area'],axis=1)
```

```python
model = sm.OLS(Y,X)
results = model.fit()
results.summary()
```

```python
pred=results.predict(X)
```

```python
def mod_metrics(act,pred):
    print("RMSE: ",np.sqrt(((act-pred)**2).mean()),"\n")
    print("MAE: ",(np.abs(act-pred).mean()),"\n")
    print("MAPE: ",(np.abs((act-pred)/act).mean()),"\n")
```

```python
mod_metrics(Y,pred)
```
