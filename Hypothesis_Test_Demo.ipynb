```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import statsmodels.api as sm
```

Paired t-test


```python
data=pd.read_excel("Hypothesis_test_demo.xlsx",sheet_name='Paired_t')
```

```python
data.head()
```

```python
stats.ttest_rel(data['Sales_pre'], data['Sales_post'],alternative='two-sided')
```

```python
help(stats.ttest_rel)
```

Indep T-Test


```python
data=pd.read_excel("Hypothesis_test_demo.xlsx",sheet_name='Indep_t')
```

```python
data.head()
```

```python
stats.ttest_ind(a=data['Test_sales'], b=data['Control_sales'], equal_var=False,alternative='two-sided')
```
